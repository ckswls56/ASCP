```python
generator_color = {'RL': 'r', 'OptaPlanner': 'g'}
solver_line_style = {'RL': '-', 'OptaPlanner': '--'}
```

```python
# 그래프 그리기
""" 
time,OptaPlanner,RL
0,14182375500,4808217185
30,	7312798513,	4204064671
60,	5123621323,	3713819785
90,	4056848663,	3352233500
120,3461735753,	3090639118

위와 같은 데이터를 가지고 그래프를 그린다고 가정하면

OptaPlanner, RL 을 각각의 그래프로 그리고 싶다면


"""
plt.figure(figsize=(20, 10))
# 여백 최대한 없애기
plt.tight_layout()
# y축 값들 log scale로 만들기
plt.yscale('log')
# 그래프 그리기
for col in dir_best_scores.columns[1:]:
    print(col)
    print(generator_color[col])
    print(solver_line_style[col])
    # 모든 행의 값이 그래프에 표시될 수 있도록
    plt.plot(dir_best_scores['time'], dir_best_scores[col], label=col, linestyle=solver_line_style[col], color=generator_color[col])

# x축 모든 행의 값이 표시될 수 있도록
plt.xticks(range(0, 601, 30), fontsize=15)
plt.yticks(fontsize=15)
plt.legend(fontsize=20)
plt.xlabel('Time(min)', fontsize=20)
plt.ylabel('Score', fontsize=20)
#plt.ylabel('Optimization Cost Score', fontsize=25)
plt.title('With RL vs Without RL using Great Deluge Algorithm', fontsize=30)
# # 그래프가 꺾이는 부분(점수가 달라지는 부분)에 점 표시 및 점수, 날짜 출력, 모든 그래프에 적용
# for dir_name, best_score_list in self.dir_best_scores.items():
#     for i, (time, score) in enumerate(best_score_list):
#         if i == 0:
#             continue
#         # 점수가 달라지는 부분에 점 표시
#         if best_score_list[i-1][1] != score:
#             plt.scatter(i, score, c='red')
#             score_src = f'{score:,}'
#             plt.text(i, score, f'{score_src}', rotation=30, verticalalignment='bottom', fontsize=7)
# plt.show()
plt.savefig('With RL vs Without RL using Great Deluge Algorithm.png')
```
